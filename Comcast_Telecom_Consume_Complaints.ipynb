```python
#df = df.set_index(df["Date"])
# convert to date
print(type(df.Date[1]))
df['Date'] = pd.to_datetime(df['Date']) #conver from str to timestamps to sort , refrence https://www.geeksforgeeks.org/how-to-sort-a-pandas-dataframe-by-date/
print(type(df.Date[1]))
df=df.sort_values(by="Date")





```

```python

df1 = df.set_index(pd.to_datetime(df["Date_month_year"] + " " + df["Time"])) # make index time include year,month,days,houres ],munits abd second #refrence https://www.geeksforgeeks.org/how-to-sort-a-pandas-dataframe-by-date/ 
df1["Date_month_year"] = pd.to_datetime(df1["Date_month_year"])
```

```python
#group by datetime ,refrence https://www.geeksforgeeks.org/how-to-group-pandas-dataframe-by-date-and-time/
#size We can use size() method to compute and display group sizes. , https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.size.html

df1.groupby(pd.Grouper(freq='M')).size().plot()

```

```python

print(df3.dtypes)
#Which complaint types are maximum i.e., around internet, network issues, or across any other domains.?
df3.sort_values(by=['frequency'],ascending=False).head(40) #order by highest complaint types
```

```python
print(df.Status.unique()) #https://pandas.pydata.org/docs/reference/api/pandas.Series.unique.html
```

```python
#a new column , refrence https://www.geeksforgeeks.org/adding-new-column-to-existing-dataframe-in-pandas/
# for Status in  df["Status"]:
    
#     if df["Status"]=="Open":
#         df["New_condation"]="Open"
#     else :
#         df["New_condation"]="Closed"
   
df["new_Status"] = ["Open" if Status=="Open" or Status=="Pending" else "Closed" for Status in df["Status"]]
df.head(5)



```

```python

df4=df.groupby(["State"]).size()  
df4=pd.DataFrame({'number of complaints':df4[:]}).sort_values(by=['number of complaints'],ascending=False)
df4.head(40) #number of complaints for each state ,, start from max
```

```python
df4.head(1) #Which state has the maximum complaints 
```

```python
# refrecne for bar chart ,, https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html
#https://stackoverflow.com/questions/17679089/pandas-dataframe-groupby-two-columns-and-get-counts
#for unstack ,,https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.unstack.html
df5=df.groupby(["State","new_Status"]).size().unstack() # Pandas DataFrame Groupby two columns and get counts ,,,, more we added unstack to make it two columns fpr Closed and Open
df5.head(10) # as you can see there are NaN ,, we should fill it with zero 
df5=df5.fillna(0)
df5
```

```python
#plot bar ,, refrence https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html 
ax = df5.plot.bar(stacked=True) 
```

```python
df5["percentage of unresolved complaints"]=100*df5["Closed"]/df5["Closed"].sum()
df5=df5.sort_values(by=['percentage of unresolved complaints'],ascending=False)
df5.head(40) #sort by max of percentage of unresolved complaints ,, so the max will be first 
```

```python
#Which state has the highest percentage of unresolved complaints
df5.head(1)
```
